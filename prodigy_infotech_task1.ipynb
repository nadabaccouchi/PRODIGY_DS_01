## Filter and Visualize Data

```python
# Filter data for total population
total_population_data = df[df['Indicator_Code'] == 'SP.POP.TOTL']

# Sort data based on the total population for 2022
total_population_sorted = total_population_data.sort_values(by="2023", ascending=False)

# Get the top ten countries with the highest total population for 2023
total_top_ten_countries = total_population_sorted.head(10)
print("Top ten countries of total population\n")
print(total_top_ten_countries[['Country_Name']] )
```

```python
# Create the bar plot
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x="2023", y="Country_Name", data=total_top_ten_countries, palette="magma")
plt.title("Top Ten Countries of Total Population (2023)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

plt.figure(figsize=(15, 6))
plt.subplot(2,2,2)
sns.barplot(x="2017", y="Country_Name", data=total_top_ten_countries, palette="magma")
plt.title("Top Ten Countries with Total Population (2017)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

```python
# Sort data based on the total population for 2023
total_population_sorted1 = total_population_data.sort_values(by="2023", ascending=True)

# Get the bottom ten countries with the highest total population for 2023
total_bottom_ten_countries = total_population_sorted1.head(10)
print("Bottom ten countries of total population\n")
print(total_bottom_ten_countries[['Country_Name']] )
```

```python
# Create the bar plot
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x="2011", y="Country_Name", data=total_bottom_ten_countries, palette="cividis")
plt.title("Bottom Ten Countries of Total Population (2011)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

plt.figure(figsize=(15, 6))
plt.subplot(2,2,2)
sns.barplot(x="2001", y="Country_Name", data=total_bottom_ten_countries, palette="cividis")
plt.title("Bottom Ten Countries with Total Population (2001)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

```python
# Filter data for male population
male_population_data = df[df["Indicator_Code"] == "SP.POP.TOTL.MA.IN"]

# Sort data based on the male population for 2023
male_population_sorted = male_population_data.sort_values(by="2023", ascending=False)

# Get the top ten countries with the highest male population for 2023
male_top_ten_countries = male_population_sorted.head(10)
print("Top ten countries of male population")
print(male_top_ten_countries[['Country_Name']] )
```

```python
# Filter data for female population
female_population_data = df[df["Indicator_Code"] == "SP.POP.TOTL.FE.IN"]

# Sort data based on the female population for 2023
female_population_sorted = female_population_data.sort_values(by="2023", ascending=False)

# Get the top ten countries with the highest female population for 2023
female_top_ten_countries = female_population_sorted.head(10)
print("Top ten countries of female population")
print(female_top_ten_countries[['Country_Name']] )
```

```python
# Create the bar plot
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x="2023", y="Country_Name", data=male_top_ten_countries, palette="crest")
plt.title("Top Ten Countries with Highest Male Population (2023)",size=10)
plt.xlabel("Male Population",size=10)
plt.ylabel("Country",size=10)
plt.show()

plt.figure(figsize=(15, 6))
plt.subplot(2,2,2)
sns.barplot(x="2023", y="Country_Name", data=female_top_ten_countries, palette="crest")
plt.title("Top Ten Countries with Highest Female Population (2023)",size=10)
plt.xlabel("Female Population",size=10)
plt.ylabel("Country",size=10)
plt.show()
```

```python
male_lowest_ten_countries = male_population_sorted.tail(10)
print("Top ten countries of lowest male population")
print(male_lowest_ten_countries[['Country_Name']] )
```

```python
female_lowest_ten_countries = female_population_sorted.tail(10)
print("Top ten countries of lowest female population")
print(female_lowest_ten_countries[['Country_Name']] )
```

```python
# Create the bar plot
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x="2023", y="Country_Name", data=male_lowest_ten_countries, palette="viridis")
plt.title("Top Ten Countries with Lowest Male Population (2023)",size=10)
plt.xlabel("Male Population",size=10)
plt.ylabel("Country",size=10)
plt.show()

plt.figure(figsize=(15, 6))
plt.subplot(2,2,2)
sns.barplot(x="2023", y="Country_Name", data=female_lowest_ten_countries, palette="viridis")
plt.title("Top Ten Countries with Lowest Female Population (2023)",size=10)
plt.xlabel("Female Population",size=10)
plt.ylabel("Country",size=10)
plt.show()
```

```python
# Merge male and female population data on 'Country_Name'
merged_data = pd.merge(male_population_data, female_population_data, on="Country_Name", suffixes=("_male", "_female"))
```

```python
merged_data
```

```python
# Calculate the total population for each country (male + female)
merged_data["Total Population"] = merged_data["2022_male"] + merged_data["2022_female"]
```

```python
merged_data.head()
```

```python
# Sort data based on total population in descending order
sorted_data = merged_data.sort_values(by="Total Population", ascending=False)

# Select the top 10 countries with the highest total population
top_10_countries = sorted_data.head(10)
```

```python
# Create the stacked bar plot
plt.figure(figsize=(12, 6))

sns.barplot(x="Country_Name", y="2023_female", data=top_10_countries, color="purple", label="Female Population")
sns.barplot(x="Country_Name", y="2023_male", data=top_10_countries, bottom=top_10_countries["2023_female"], color="pink", label="Male Population")
plt.title("Top 10 Countries with Male and Female Populations (2023)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# Select the bottom 10 countries with the highest total population
bottom_10_countries = sorted_data.tail(10)
```

```python
# Create the stacked bar plot
plt.figure(figsize=(12, 6))

sns.barplot(x="Country_Name", y="2023_female", data=bottom_10_countries, color="orange", label="Female Population")
sns.barplot(x="Country_Name", y="2023_male", data=bottom_10_countries, bottom=bottom_10_countries["2023_female"], color="green", label="Male Population")
plt.title("Bottom 10 Countries with Male and Female Populations (2023)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45, ha="right")
plt.show()
```
